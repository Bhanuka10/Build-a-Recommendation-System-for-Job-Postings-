```python
import pandas as pd
import numpy as np
```

```python
candidates = pd.read_csv('../data/raw/job_recommendation_dataset.csv')
jobs       = pd.read_csv('../data/raw/resume_dataset_2.csv')

print("Candidates shape:", candidates.shape)
print("Jobs shape:", jobs.shape)
print("\n── Candidate columns ──")
print(candidates.columns.tolist())
print("\n── Job columns ──")
print(jobs.columns.tolist())
```

```python
# ── Basic info ─────────────────────────────────────────────────
print("\n── Missing values in candidates ──")
print(candidates.isnull().sum())

print("\n── Missing values in jobs ──")
print(jobs.isnull().sum())

# ── Preview ────────────────────────────────────────────────────
display(candidates.head(3))
display(jobs.head(3))

# ── Save copies to processed folder ───────────────────────────
candidates.to_csv('../data/processed/candidates_raw_copy.csv', index=False)
jobs.to_csv('../data/processed/jobs_raw_copy.csv', index=False)
print("\nRaw copies saved to data/processed/")
```
